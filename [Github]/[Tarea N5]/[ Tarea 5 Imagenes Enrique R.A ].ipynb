```python
#########################################################################################################################################################################################################################################################################
s = np.linspace(-1,1,1000)
x_l = np.linspace(-1,1,1000)
t = np.linspace(0,10/3310,1000)
x_t = np.linspace(-3,3,1000)
x_l, t = np.meshgrid(x_l,t)
c1=331
#########################################################################################################################################################################################################################################################################
Fi=(1-abs(s))
p_p=(1-abs(x_l + t*c1))
p_n=(1-abs(x_l - t*c1))
#########################################################################################################################################################################################################################################################################
def p_p_t(T):
  ti=5/3310
  c1=331 # velocidad del sonido en el aire
  if abs(T+ti*c1)<=1:
    return (1-abs(T+ti*c1))
  else: 
    return 0
#########################################################################################################################################################################################################################################################################
def p_n_t(T):
  ti=5/3310
  c1=331 # velocidad del sonido en el aire
  if abs(T-ti*c1)<=1:
    return (1-abs(T-ti*c1))
  else: 
    return 0
#########################################################################################################################################################################################################################################################################
lista_2=[p_n,p_p]
lista_c=["Red","Blue"]
#########################################################################################################################################################################################################################################################################
plt.figure(figsize=(6,5))
plt.title(f"INTUICION DE LA FUNCION $\phi$", fontsize=16)
plt.plot(Fi,color="Purple")
plt.show()
#########################################################################################################################################################################################################################################################################
plt.figure(figsize=(6,5))
plt.title(f"Grafico $P_-$ para $T_0$ fijo ", fontsize=16)
plt.ylim(-0.1,1.2)
plt.xlim(-1.9,0.9)
plt.plot(x_t, [p_p_t(i) for i in x_t],color="Blue",linewidth=2)
plt.plot(0.6,0, ">",color="Blue")
plt.plot(0.7,0, ">",color="Blue")
plt.plot(0.8,0, ">",color="Blue")
plt.plot(-1.6,0, ">",color="Blue")
plt.plot(-1.7,0, ">",color="Blue")
plt.plot(-1.8,0, ">",color="Blue")
plt.legend(["$P_+$"])
plt.show()
#########################################################################################################################################################################################################################################################################
plt.figure(figsize=(6,5))
plt.title(f"Grafico $P_+$ para $T_0$ fijo", fontsize=16)
plt.ylim(-0.1,1.2)
plt.xlim(-0.9,1.9)
plt.plot(x_t, [p_n_t(i) for i in x_t],color="Red",linewidth=2)
plt.plot(-0.6,0,   "<",color="Red")
plt.plot(-0.7,0,   "<",color="Red")
plt.plot(-0.8,0,   "<",color="Red")
plt.plot(1.6,0,  "<",color="Red")
plt.plot(1.7,0,  "<",color="Red")
plt.plot(1.8,0,  "<",color="Red")
plt.legend(["$P_-$"])
plt.show()
#########################################################################################################################################################################################################################################################################
plt.figure(figsize=(6,5))
plt.title(f"Grafico $P_+$ y $P_-$ para $T_0$ fijo", fontsize=16)
plt.ylim(-0.1,1.2)
plt.xlim(-2,2)
plt.plot(x_t, [p_p_t(i) for i in x_t],color="Blue",linewidth=2)
plt.plot(x_t, [p_n_t(i) for i in x_t],color="Red",linewidth=2)
plt.legend(["$P_+$","$P_-$"]) 
plt.plot(-0.8,0,   "<",color="Red")
plt.plot(-0.9,0,   "<",color="Red")
plt.plot(-1,0,   "<",color="Red")
plt.plot(1.6,0,  "<",color="Red")
plt.plot(1.7,0,  "<",color="Red")
plt.plot(1.8,0,  "<",color="Red")
plt.plot(0.8,0, ">",color="Blue")
plt.plot(0.9,0, ">",color="Blue")
plt.plot(1,0, ">",color="Blue")
plt.plot(-1.6,0, ">",color="Blue")
plt.plot(-1.7,0, ">",color="Blue")
plt.plot(-1.8,0, ">",color="Blue")
plt.show()
#########################################################################################################################################################################################################################################################################
```

```python
def p_p_tt(T,ti):
  c1=331 # velocidad del sonido en el aire
  if abs(T+ti*c1)<=1:
    return (1-abs(T+ti*c1))
  else: 
    return 0
#########################################################################################################################################################################################################################################################################
def p_n_tt(T,ti):
  c1=331 # velocidad del sonido en el aire
  if abs(T-ti*c1)<=1:
    return (1-abs(T-ti*c1))
  else: 
    return 0

lista=[]
ti=1/33100
while ti<=4/331:
  lista.append(ti)
  ti+=1/33100

for e in lista:
  
  plt.title(f"Grafico movil $P_+$ y $P_-$ para $T_0$ fijo", fontsize=16)
  plt.ylim(-0.1,1.2)
  plt.xlim(-3,3)
  plt.plot(x_t, [p_p_tt(i,e) for i in x_t],color="Blue",linewidth=2)
  plt.plot(x_t, [p_n_tt(i,e) for i in x_t],color="Red",linewidth=2)
  plt.legend(["$P_+$","$P_-$"]) 
  plt.show()
  clear_output(wait = True)
```
